```r
library(ggplot2)

#Set the width and of the plot 
options(repr.plot.width=8, repr.plot.height=4)

options(warn=-1)
```

```r
#setwd("/Users/Data")
#getwd()
data.Main <- read.csv(url("https://raw.githubusercontent.com/sqlshep/SQLShepBlog/master/data/USA.dataAll.csv"),stringsAsFactors=FALSE,header=TRUE)

#remove all NAs and infinites 
is.na(data.Main) <- sapply(data.Main, is.infinite)
data.Main[is.na(data.Main)] <-0

data.Main <- data.Main[!(data.Main$X == "1864"),]
```

##### Notice the nice curve in the line, this is what we are looking for to be a typically good varialbe for an logistic regression.  It just so happens if you look back on the eleciton, counties that were heavily populated, which typially contained large cities did not vote for Trump, so the data supports that. 

```r
ggplot(data.Main, aes(lg_Population, Winner,colour=Winner)) +
  stat_smooth(method="glm", method.args=list(family="binomial"), se=FALSE)+
  geom_point() +
  ggtitle("Likelihood of Voting for Trump")
```
