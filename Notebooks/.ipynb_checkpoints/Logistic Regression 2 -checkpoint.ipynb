```r
#setwd("/Users/Data")
#getwd()
data.Main <- read.csv(url("https://raw.githubusercontent.com/sqlshep/SQLShepBlog/master/data/USA.dataAll.csv"),stringsAsFactors=FALSE,header=TRUE)
```

```r
head(data.Main)
```

```r
library(car)
library(choroplethr)
library(choroplethrMaps)
library(ggplot2)

options(scipen=999)
#TURN WARNINGS OFF
options(warn=-1)
#TURN WARNINGS ON
#options(warn=0)

```

```r
#remove all NAs and infinites 
is.na(data.Main) <- sapply(data.Main, is.infinite)
data.Main[is.na(data.Main)] <-0

data.Main <- data.Main[!(data.Main$X == "1864"),]
```

```r
names(data.Main[6:26])
#names(data.Main)
```

```r
#Set the width and of the plot 
options(repr.plot.width=8, repr.plot.height=4)

options(warn=-1)
```

```r
ggplot(data.Main, aes(lg_Population, Winner,colour=Winner)) +
  stat_smooth(method="glm", method.args=list(family="binomial"), se=FALSE)+
  geom_point() +
  ggtitle("Likelihood of Voting for Trump")
```

```r
ggplot(data.Main, aes(Population, Winner,colour=Winner)) +
  stat_smooth(method="glm", method.args=list(family="binomial"), se=FALSE)+
  geom_point() +
  geom_text(data=subset(data.Main, Population > 3500000),aes(Population,Winner,label=county_name), angle=90,hjust=-.1,) +
  ggtitle("Likelihood of Voting for Trump")
```

```r
ggplot(data=data.Main, aes(Population)) + 
  geom_histogram(aes(fill=..count..))
```

```r
ggplot(data=data.Main, aes(lg_Population)) + 
  geom_histogram(aes(fill=..count..))
```
