```python
import nltk
import os
import string

import matplotlib.pyplot as plt


from wordcloud import WordCloud


from nltk.tokenize import RegexpTokenizer
from collections import Counter

nltk.download("stopwords")
nltk.download("punkt")
```

```python

stop_words = nltk.corpus.stopwords.words("english") + list(string.punctuation) \
+ list(string.ascii_lowercase) + list(string.ascii_uppercase) + list(string.digits) \
+ list(['--']+["''"]+["``"]+[".."]+["..."]+["ii"]+["iii"]+["iv"]+["'s"])

```

```python
os.chdir("/Users/Me/Data")
os.getcwd()

a = open("HR1.txt") 
HR1 = a.read()

HR1 = HR1.lower()

HR1 =  HR1.replace("`","")
HR1 =  HR1.replace("'","")
```

```python
HR1_token = nltk.word_tokenize(HR1)

hr1_filter = [w for w in HR1_token if not w in stop_words]

#Same

#filtered_sentence = []
#for w in HR1_token:
#    if w not in stop_words:
#        filtered_sentence.append(w)


hr1_counter = Counter(hr1_filter)
```

```python
sorted_word_counts = sorted(list(hr1_counter.values()), reverse=True)

plt.loglog(sorted_word_counts)
plt.ylabel("Freq")
plt.xlabel("Word Rank");
```

```python
plt.hist(sorted_word_counts, bins=50);
```
